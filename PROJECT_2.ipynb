```python
'''In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter 
and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement 
CRUD( create, read, update,delete) operations on a ‘books’ database, in SQLite.

The window will contain text widgets to accept/ display book title, author, year of publication, 
ISBN (optional) and a large text area/ list box to display the records which match the search criteria. 
It should contain buttons to perform the following operations on the ‘books’ database:

 
display all the records
search a book
add a book
issue a book
delete a book
exit the application
 

When option 1 is selected, all the books present in the library should be displayed. 
Option 2 should be able to search the book based on any of the criteria like author name, book title, 
year or ISBN. Similarly, it should be able to add the entry when option 3 is selected and issue a book when option 4 i
s selected. The book details should be deleted when option 5 is selected. Selecting option 6 should provide a graceful 
exit from the app.'''

from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMassgeBox

root=Tk()
root.title('LIBRARY BOOK MANAGEMENT')

def Database():
    global conn,cursor
    conn=sqlite3.connect('LIBRARY.db')
    cursor=conn.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS container(authourname TEXT, booktitle TEXT, yearofpublication INTEGER,isbn INTEGER UNIQUE)')
    
def add_new_book():
    if str(authourname.get(1.0,'end-1c'))=="" or str(booktitle.get(1.0,'end-1c'))=="" or str(yearofpublication.get(1.0,'end-1c'))=="" or str(isbn.get(1.0,'end-1c'))=="":
        txt_result.config(text="Please enter all the fields",fg="red")
    else:
        Database()
        cursor.execute('INSERT INTO container(authourname,booktitle,yearofpublication) values(?,?,?,?)',(authourname.get(1.0,'end-1c'),booktitle.get(1.0,'end-1c'),yearofpublication.get(1.0,'end-1c'),isbn.get(1.0,'end-1c')))
        conn.commit()
        authourname.delete("1.0","end")                                                                                                                              
        booktitle.delete("1.0","end")
        yearofpublication.delete("1.0","end")
        isbn.delete("1.0","end")
        cursor.close()
        conn.close()
        txt_result.config(text="successfully book details added in library !",fg="green")
        Display_function()
                                                                                                                                      
def Display_function():
    Database()
    tree.delete(*tree.get_children())
    cursor.execute("SELECT * From container ORDER BY booktitle ASC")
    fetch=cursor.fetchall()
    for data in fetch:
        tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
    conn.close()
                                                                                                                                      
def Display_all_records():
    Display_function()
    txt_result.config(text="!see all books available in library",fg="green")
    
def Issue_a_book():
    Database()
    if isbn.get(1.0,'end-1c')=="":
        txt_result.config(text="please enter ISBN numnber to issue",fg="red")
    else:
        a=str(isbn.get(1.0,'end-1c'))
        isbn.delete("1.0","end")
        l.append(a)
        for i in l:
            temp="DELETE FROM container where isbn =  "+i
            cursor.execute(temp)
        txt_result.config(text="successfully Issued!",fg="green")
        
        tree.delete(*tree.get_children())
        cursor.execute("SELECT * From container ORDER BY booktitle ASC")
        fetch=cursor.fetchall()
        for data in fetch:
            tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
        conn.close()
                                                                                                                                      
def search_a_book():
    global temp
    
    def display():
        booktitle.delete("1.0","end")                                                                                                                              
        authourname.delete("1.0","end")
        yearofpublication.delete("1.0","end")
        isbn.delete("1.0","end")
        tree.delete(*tree.get_children())
        Database()
        cursor.execute(temp)
        fetch=cursor.fetchall()
        if fetch:
            for data in fetch:
                tree.insert("",'end',values=(data[0],data[1],data[2],data[3]))
            txt_result.config(text="successfully fetched data",fg="green")
        else:
            txt_result.config(text="Book details not founded.please give valid data",fg="red")
        conn.close()
    if str(booktitle.get(1.0,'end-1c'))!="":
        temp1="'"+str(booktitle.get(1.0,'end-1c'))+"'"
        temp="SELECT * FROM container WHERE booktitle = "+temp1
        display()
    elif str(authourname.get(1.0,'end-1c'))!="":
        temp1="'"+str(authourname.get(1.0,'end-1c'))+"'"
        temp="SELECT * FROM container WHERE authourname = "+temp1
        display()
    elif str(yearofpublication.get(1.0,'end-1c'))!="":
        temp1="'"+str(yearofpublication.get(1.0,'end-1c'))+"'"
        temp="SELECT * FROM container WHERE yearofpublication = "+temp1
        display()
    elif str(isbn.get(1.0,'end-1c'))!="":
        temp1="'"+str(isbn.get(1.0,'end-1c'))+"'"
        temp="SELECT * FROM container WHERE isbn = "+temp1
        display()
    else:
        txt_result.config(text="please fill atleast one field to search book details ! ",fg="red")
                                                                                                                                      
def Delete():
    if booktitle.get(1.0,"end-1c")!="":
        text_result.config(text="Please enter booktitle!",fg="red")
    else:
        temp1="'"+str(booktitle.get(1.0,"end-1c"))+"'"
        temp="DELETE FROM container WHERE booktitle = "+temp1
        Database()
        cursor.execute(temp)
        conn.commit()
        cursor.close()
        text_result.config(text="sucessfully deleted  book data",fg='red')
        Display_function()
                                                                                                                                             
    
def Exit():
    result=tkMassgeBox.askquestion("Do you want to exit? (y/n)",icon="warning")
    if result =="yes":
        root.destroy()
        exit()


Top=Frame(root,width=800,height=50,bd=8,relief='raise')
Top.pack(side=TOP)
Left=Frame(root,width=600,height=500,bd=8,relief='raise')
Left.pack(side=LEFT)
Right=Frame(root,width=800,height=900,bd=8,relief='raise')
Right.pack(side=RIGHT)

Forms=Frame(Left,width=600,height=550)
Forms.pack(side=TOP)
Buttons=Frame(Left,width=300,height=100,bd=10,relief='raise')
Buttons.pack(side=BOTTOM)


txt_title=Label(Top,width=900,font=('arial',24),text="library")
txt_title.pack()
txt_authourname=Label(Forms,text="AUTHOUR NAME:",font=('arial',16),bd=15)
txt_authourname.grid(row=1,stick='e')
txt_booktitle=Label(Forms,text="BOOK TITLE:",font=('arial',16),bd=15)
txt_booktitle.grid(row=0,stick='e')
txt_yearofpublication=Label(Forms,text="YEAR OF PUBLICATION:",font=('arial',16),bd=15)
txt_yearofpublication.grid(row=2,stick='e')
txt_isbn=Label(Forms,text="ISBN:",font=('arial',16),bd=15)
txt_isbn.grid(row=3,stick='e')
txt_result=Label(Buttons)
txt_result.pack(side=TOP)

booktitle=Text(Forms,height=2.5,width=30)
booktitle.grid(row=0,column=1)
authourname=Text(Forms,height=2.5,width=30)
authourname.grid(row=1,column=1)
yearofpublication=Text(Forms,height=2.5,width=30)
yearofpublication.grid(row=2,column=1)
isbn=Text(Forms,height=2.5,width=30)
isbn.grid(row=3,column=1)



btn_update=Button(Buttons,width=10,text="option1",command=Display_all_records)
btn_update.pack(side=LEFT)
btn_read=Button(Buttons,width=10,text="option2",command=search_a_book)
btn_read.pack(side=LEFT)
btn_create=Button(Buttons,width=10,text="option3",command=add_new_book)
btn_create.pack(side=LEFT)
global l
l=[]
btn_delete=Button(Buttons,width=10,text="option4",command=Issue_a_book)
btn_delete.pack(side=LEFT)
btn_delete=Button(Buttons,width=10,text="option5",command=Delete)
btn_delete.pack(side=LEFT)
btn_exit=Button(Buttons,width=10,text="option6",command=Exit)
btn_exit.pack(side=LEFT)


scrollbary=Scrollbar(Right,orient=VERTICAL)
scrollbarx=Scrollbar(Right,orient=HORIZONTAL)

tree=ttk.Treeview(Right,columns=("authourname","booktitle","yearofpublication","ISBN"))
scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM,fill=X)



tree.heading("authourname",text="authourname",anchor=W)
tree.heading("booktitle",text="booktitle",anchor=W)
tree.heading("yearofpublication",text="yearofpublication",anchor=W)
tree.heading("ISBN",text="ISBN",anchor=W)

tree.column("#0",minwidth=0,width=0)
tree.column("#1",minwidth=0,width=130)
tree.column("#2",minwidth=0,width=135)
tree.column("#3",minwidth=0,width=140)
tree.pack()


root.mainloop()


```
